## Importações

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

'''o set_options faz o pandas exibir todas as colunas de um DataFrame
e o set do seaborn define o estilo dos gráficos para 'whitegrid'
'''

pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')
```

## Carregar Dados

```python
# aqui eu usei a função read_csv para ler e buscar o arquivo bruto do covid_data.csv
# e transforma ele em um dataframe 
# o head() vai mostrar as 5 primeiras linnhas do dataframe
 
df = pd.read_csv('../data/raw/covid_data.csv')
df['date'] = pd.to_datetime(df['date'])

df.head()
```

## Informações basica

```python
#info vai mostrar a quantidade de colunas o numero de linhas de cada coluna e o seus tipo de dados
#describe mostra estaticas basicas do datafreme como procentagens,medias, maximos, minimus e etc
df.info()
df.describe()
```

## Verificar uma colouna

```python
df.columns
df['location'].unique()[:10]

''''vai mostrar apenas as informações da coluna location, a função unique()
vai mostrar apenas valores unicos ou seja se houver 5 palavras "Brazil" vai
mostrar apenas uma e o [:10] seve para delimitar a quantidade de informações mostradas '''
```

## Exploração visual simples

```python
brasil = df[df['location'] == 'Brazil']

plt.figure(figsize=(12,6))
sns.lineplot(data=brasil, x='date', y='total_cases', color='red')
plt.title("Evolução dos Casos Totais de COVID-19 no Brasil", fontsize=14)
plt.xlabel("Data")
plt.ylabel("Total de Casos")
plt.grid(True)
plt.show()
```
